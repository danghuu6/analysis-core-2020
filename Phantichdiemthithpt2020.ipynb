### Xem và vẽ biểu đồ thống kê của kiểu dữ liệu điểm môn Vật lý

```python
# Xem kiểu dữ liệu điểm của môn Vật lí
type_float = 0
type_str = 0
type_other = 0
for i in data['Vatli']:
  if type(i) == float:
    type_float += 1
  elif type(i) == str:
    type_str += 1
  else:
    type_other += 1
```

```python
# Biểu đồ hiển thị các kiểu dữ liệu của môm vật lí
data_type = {"Float": type_float,
             "Str": type_str,
             "Other": type_other}
plt.bar(data_type.keys(), data_type.values())

# Thêm chỉ số trên mỗi cột
for x, y in zip(data_type.keys(), data_type.values()):
  plt.text(x, y, '%d'%y, ha = 'center', va= 'bottom')

plt.savefig('kieudulieuvatli')
plt.show()
```

### Đổi kiểu dữ liệu str về float, xử lý ngoại lệ 2 giá trị không xác định gán bằng none

```python
# Đổi kiểu str về float và bắt ngoại lệ 2 giá trị không xác định đưa về None của môn Vật lí
for i in range(len(data['Vatli'])):
  try:
    if type(data['Vatli'][i]) == str:
      data['Vatli'][i] = float(data['Vatli'][i])
  except:
    data['Vatli'][i] = np.nan
```

### Thống kê khoảng điểm từng môn và số lượng của nó

```python
# Hàm thống kê khoảng điểm
def thongkediem(dt, mon):
  diem01 = 0 # điểm dưới 1
  diem14 = 0 # từ 1 đến 4
  diem46 = 0 # từ 4 đến 6
  diem68 = 0 # từ 6 đến 8
  diem810 = 0 # từ 8 đến dưới 10
  diem10 = 0 # điểm 10
  for i in dt[mon]:
    if i <= 1:
      diem01 += 1
    elif i > 1 and i < 4:
      diem14 += 1
    elif i >= 4 and i < 6:
      diem46 += 1
    elif i >= 6 and i < 8:
      diem68 += 1
    elif i >= 8 and i < 10:
      diem810 += 1
    elif i == 10:
      diem10 += 1
  return [diem01, diem14, diem46, diem68, diem810, diem10]
```

```python
def thongke(dt, mon):
  dt_null = pd.DataFrame(None, columns=['Khoang diem','So luong'])
  lst1 = ['Duoi 1', 'hon 1 -> <4','4 -> <6','6 -> <8','8 -> 10','Diem 10']
  dt_null['Khoang diem'] = lst1
  dt_null['So luong'] = thongkediem(dt, mon)
  return dt_null
```

```python
# Thông kê từng môn

for i in dt_new.drop(dt_new.columns[[3,4]], axis=1):
  print('Thống kê điểm',i)
  print(thongke(dt_new, i))
```

### Lặp bảng thống kê thông tin điểm của 9 môn và 2 tổ hợp

```python
#Tạo data rỗng
bangthongke = pd.DataFrame(None,columns=['Môn','Tổng số thí sinh','Điểm trung bình','So thí sinh liệt(<=1)','Số thí sinh dưới trung bình(<5)','Điểm có nhiều thí sinh đạt nhất'])
#Tạo các List tiêu đề
list1 = [] # Danh sách các môn học
list2 = [] # Tổng số thí sinh
list3 = [] # Điểm trung bình của tổng các thí sinh
list4 = [] # Số thí sinh bị liệt
list5 = [] # Số thí sinh dưới trung bình(<5)
list6 = [] # Điểm có nhiều thí sinh đạt nhất

for i in dt_new.columns:
  list1.append(i) #Lấy tên mỗi môn
  tongso = dt_new[i].dropna().count() #Đếm số học sinh theo môn học
  tongdiem = dt_new[i].sum() #Cộng điểm của toàn bộ học sinh theo môn học
  trungbinh = tongdiem/tongso #Tính trung bình mỗi môn
  count1 = 0
  count2 = 0
  for j in dt_new[i]:
    if j < 5:
      count1 += 1 #Đếm số học sinh nhỏ hơn 5 điểm của mỗi môn
    if j <= 1:
      count2 += 1 #Đếm số học sinh bị điểm liệt của từng môn

  # nhohon5diem=count1/tongso
  value_max = dt_new[i].value_counts().max() #Tìm đểm có nhiều thí sinh đạt được
  diem_mon = dt_new[i].value_counts().sort_index() #Đếm số học sinh theo môn học
  df_dtnew = pd.DataFrame(diem_mon) #Tạo bảng data
  df_dtnew = df_dtnew.transpose() #Chuyển hàng thành cột và ngược lại

  #Tìm điểm trong bảng data có nhiều thí sinh đạt được
  for n in range(0,len(df_dtnew.columns)):
    if df_dtnew.iloc[0,n]==value_max:
      break

  value_max2 = diem_mon.index[n]
  #Lưu các thông số vào array
  list2.append(tongso)
  list3.append(trungbinh)
  list4.append(count2)
  list5.append(count1)
  list6.append(value_max2)
  
#Lưu các aray vào bảng
bangthongke['Môn'] = list1
bangthongke['Tổng số thí sinh'] = list2
bangthongke['Điểm trung bình'] = list3
bangthongke['So thí sinh liệt(<=1)'] = list4
bangthongke['Số thí sinh dưới trung bình(<5)'] = list5
bangthongke['Điểm có nhiều thí sinh đạt nhất'] = list6

bangthongke
```
